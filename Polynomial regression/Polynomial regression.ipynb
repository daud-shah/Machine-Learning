# Polynomial Regression in Machine Learning

Polynomial regression is an extension of linear regression used when the relationship between the independent variable \( x \) and the dependent variable \( y \) is nonlinear. Instead of fitting a straight line (as in linear regression), it fits a curve to the data by including higher-degree terms of the independent variable.

## Key Concepts of Polynomial Regression

1. **Model Representation**:
   The model represents the relationship between \( x \) and \( y \) as:
   \[
   y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + \cdots + \beta_nx^n + \epsilon
   \]
   Where:
   - \( \beta_0, \beta_1, \beta_2, \ldots, \beta_n \) are the coefficients.
   - \( n \) is the degree of the polynomial.
   - \( \epsilon \) is the error term.

2. **Linear vs. Polynomial Regression**:
   - Polynomial regression is still a **linear model** in the sense that the coefficients (\( \beta \)) are linear, but the input features (powers of \( x \)) make the relationship nonlinear.

3. **Feature Transformation**:
   The input feature \( x \) is transformed into \( x, x^2, x^3, \ldots, x^n \). This allows the model to capture more complex patterns in the data.

4. **Choosing the Degree of the Polynomial**:
   - A higher-degree polynomial can model more complex relationships.
   - **Underfitting** occurs when the degree is too low, failing to capture the data's complexity.
   - **Overfitting** occurs when the degree is too high, leading to a model that fits noise in the data.

## Steps to Implement Polynomial Regression in Python

```python
#Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Example dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1.5, 2.8, 4.6, 8.1, 10.2])
```

```python
poly = PolynomialFeatures(degree=2)  # Degree of polynomial
X_poly = poly.fit_transform(X)      # Transform features
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)
```

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Polynomial Fit')
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression')
plt.show()
```

```python
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```
