### Project: Predicting House Prices Based on Area
     Problem Statement:
      We have a dataset containing the area (in square feet) of houses and their corresponding prices. The relationship between area and price is nonlinear, and weâ€™ll use polynomial regression to model it.
-----------
#### Steps to Solve:
    Generate a Synthetic Dataset.
    Visualize the Data.
    Apply Polynomial Regression.
    Evaluate the Model.
    Visualize the Results.

```python
#  Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Generate synthetic data
np.random.seed(42)  # For reproducibility
area = np.linspace(500, 3000, 30).reshape(-1, 1)  # Areas (square feet)
price = 50 + 0.2 * area + 0.0001 * area**2 + 5000 * np.sin(area / 500) + np.random.normal(0, 1000, size=area.shape)  # Nonlinear prices

# Visualize the data
plt.scatter(area, price, color='blue', label='Data Points')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('House Prices vs Area')
plt.legend()
plt.show()
```

```python
# Apply Polynomial Regression
# Transform features
degree = 3  # Set the degree of the polynomial
poly = PolynomialFeatures(degree=degree)
area_poly = poly.fit_transform(area)

# Train the model
model = LinearRegression()
model.fit(area_poly, price)
```

```python
# Predict prices
price_pred = model.predict(area_poly)

# Evaluate the model
mse = mean_squared_error(price, price_pred)
print(f"Mean Squared Error: {mse}")
```

```python
# Plot the results
plt.scatter(area, price, color='blue', label='Actual Data')
plt.plot(area, price_pred, color='red', label=f'Polynomial Fit (Degree {degree})')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()
```
